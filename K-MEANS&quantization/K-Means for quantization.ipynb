```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
img = cv.imread(r'D:\Anaconda project\K-MEANS&quantization\test.jpg')
#original image
plt.imshow(img)
plt.show()
plt.close()
z = img.reshape((-1, 3))
z = np.float32(z)
criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 10, 1.0)
#################
k1 = 4
ret1, label1, center1 = cv.kmeans(z, k1, None, criteria, 10, cv.KMEANS_RANDOM_CENTERS)
center1 = cv.convertScaleAbs(center1)
pic1 = center1[label1.flatten()]
res = pic1.reshape((img.shape))
plt.imshow(res)
plt.show()
plt.close()
#################
k2 = 8
ret2, label2, center2 = cv.kmeans(z, k2, None, criteria, 10, cv.KMEANS_RANDOM_CENTERS)
center2 = cv.convertScaleAbs(center2)
pic2 = center2[label2.flatten()]
res = pic2.reshape((img.shape))
plt.imshow(res)
plt.show()
plt.close()
#################
k3 = 16
ret3, label3, center3 = cv.kmeans(z, k3, None, criteria, 10, cv.KMEANS_RANDOM_CENTERS)
center3 = cv.convertScaleAbs(center3)
pic3 = center3[label3.flatten()]
res = pic3.reshape((img.shape))
plt.imshow(res)
plt.show()
plt.close()
```
