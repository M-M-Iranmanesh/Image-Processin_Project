```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
img = cv.imread(r'D:\Anaconda project\Afne  transformations\Afne  transformations\test.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
plt.close()
```

```python
height, width = img.shape[:2]
#Scaling 3x
scaledImg = cv.resize(img, (3*height, 3*width))
plt.imshow(scaledImg)
plt.show()
plt.close()
```

```python
#Rotating 67
center = (width/2, height/2)
rotationMatrix = cv.getRotationMatrix2D(center, 67, 1)
rotationImg = cv.warpAffine(img, rotationMatrix, (width, height))
plt.imshow(rotationImg)
plt.show()
plt.close()
```

```python
#Translating
translationMatrix = np.float32([[1, 0, 132], [0, 1, 68]])
translationImg = cv.warpAffine(img, translationMatrix, (width, height))
plt.imshow(translationImg)
plt.show()
plt.close()
```

```python
#Shearing
M2 = np.float32([[1, 0, 0], [0.4, 1, 0]])
M2[0, 2] = -M2[0, 1]* width/2
M2[1, 2] = -M2[1, 0]* height/2
shearingImg = cv.warpAffine(img, M2, (width, height))
plt.imshow(shearingImg)
plt.show()
plt.close()
```
